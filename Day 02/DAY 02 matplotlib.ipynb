```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years=[1,1000,1500,1600,1700,1750,1800,1850,1900,1950,1955,1960,1965,1970,1980,1985,1990,
       1995,2000,2005,2010,2015]
pops=[200,400,458,580,682,791,1000,1262,1650,2525,2758,3018,3322,3682,
      4061,4440,4853,5310,5735,6127,6520,7349]
plt.xlabel("year")
plt.ylabel("pops")
plt.title("Population Growth")
plt.plot(years, pops)
plt.show()
```

```python
x = [1,6,3]
y = [5,9,4]

x2 = [1,2,3]
y2 = [10,14,12]
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title("it is a title")
lines = plt.plot(x, y, label="first graph")
plt.plot(x2, y2, label="second graph")
plt.legend()
plt.setp(lines,marker='*')
plt.grid(True)
plt.show()
```

```python
# sin 
x = np.arange(0, 10, 0.0001)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.plot(x, y_sin, label="sin graph")
plt.plot(x, y_cos, label="cos graph")
plt.legend()
```

```python
labels = ['python', 'c', 'c++', 'PHP', 'Java', 'Ruby']
votes = [33, 52, 12, 17, 42, 48]
```

```python
plt.pie(votes, labels=labels, autopct="%.2f%%", explode=(0.1,0,0,0,0,0))
plt.show()
```

```python
x = np.arange(1, 10)
y = np.arange(11,20)
y2 = np.arange(31,40)
plt.scatter(x, y)
plt.scatter(x, y2)
plt.show()
```

```python
korea_scores=(554,536,538)
canada_scores=(518,523,525)
china_scores=(413,570,580)
franch_scores=(495,505,499)
index=np.arange(3)
bar_width=.2
k1=plt.bar(index,korea_scores,bar_width,label="Korea")
c1=plt.bar(index+bar_width,canada_scores,bar_width,label="Canada")
ch1=plt.bar(index+bar_width*2,china_scores,bar_width,label="China")
f1=plt.bar(index+bar_width*3,franch_scores,bar_width,label="Franch")
plt.xticks(index+.6/2,('Mathematics','Reading','Science'))
plt.ylabel('Mean score in PISA in 2012')
plt.xlabel('Subjects')
plt.title('Test scores by Country')
plt.grid(True)
plt.legend()
plt.show()
```

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]
plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nStack Plots')
plt.show()
```
