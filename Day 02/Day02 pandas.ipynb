```python
mean_value = df.points.mean()
print(mean_value)
```

```python
df.points.map(lambda p:p-mean_value)
```

```python
df.points
```

```python
# 87.8884184721394
df.points.map(lambda p:p-mean_value)
```

```python
def remean_points(row):
    row.points = row.points - mean_value
    return row

df.apply(remean_points, axis='columns') #index
```
